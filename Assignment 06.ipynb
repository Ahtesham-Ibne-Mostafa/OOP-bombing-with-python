```python
# task 1

class Student:
    ID = 0
    
    def __init__(self, name, department, age, cgpa):
        self.name = name
        self.department = department
        self.age = age
        self.cgpa = cgpa
        Student.ID += 1
        
    def get_details(self):
        print("ID:", Student.ID)
        print("Name:", self.name)
        print("Department:", self.department)
        print("Age:", self.age)
        print("CGPA:", self.cgpa)
        
    @classmethod
    def from_String(cls, strng):
        lst = strng.split("-")
        objct = Student(lst[0],lst[1],lst[2],lst[3])
        return objct
     
s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details() 
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()

```

```python
# task 2

class Assassin:
    count = 0
    def __init__(self, name, rate):
        self.name = name
        self.rate = rate
        Assassin.count += 1
        
    @classmethod
    def failureRate(cls, name, rate):
        objct = cls(name, (100-rate))
        return objct
    
    @classmethod
    def failurePercentage(cls, name, rate):
        return cls(name, ('100'-rate[-1]))
    
    def printDetails(self):
        print("Name:", self.name)
        print("Success rate: {}%".format(self.rate))
        print("Total number of Assassin:", Assassin.count)



john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', '10%')
akabane.printDetails()
```

```python
#task 3

class Passenger:
    count = 0
    
    def __init__(self, name):
        self.name = name
        self.base_fare = 450
        Passenger.count += 1
        
    def set_bag_weight(self, weight):
        if weight>=21 and weight<=50:
            self.base_fare = self.base_fare + 50
            return self.base_fare
        elif weight>50:
            self.base_fare = self.base_fare + 100
            return self.base_fare
        else:
            pass
        
    def printDetail(self):
        print("Name:", self.name)
        print("Bus Fare:", self.base_fare)



print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
# task 4

class Travel:
    count = 0
    
    def __init__(self, source, destination):
        self.source = source
        self.destination = destination
        self.time = "1:00"
        Travel.count += 1
        
    def set_time(self, time):
        self.time = "{}:00".format(time)
        
    def set_source(self, source):
        self.source = source
        
    def set_destination(self, destination):
        self.destination = destination
        
    def display_travel_info(self):
        return "Source:{} \nDestination:{} \nFlight Time:{}".format(self.source,self.destination,self.time)
        


print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka","India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur","Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka","New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka","India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller =", Travel.count)
```

```python
# task 5

class Employee:
    def __init__(self, name, workingPeriod):
        self.name = name
        self.workingPeriod = workingPeriod
        
    @classmethod
    def employeeByJoiningYear(cls, name, year):
        return cls(name, (2022-year))
    
    @staticmethod
    def experienceCheck(workingPeriod, gender):
        if workingPeriod < 3:
            if gender == "male":
                return "He is not experienced"
            elif gender == "female":
                return "She is not experienced"
        else:
            if gender == "male":
                return "He is experienced"
            elif gender == "female":
                return "She is experienced"

employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
# tsak 6

class Laptop:
    laptopCount = 0
    def __init__(self, name, count):
        self.name = name
        self.count =  count
        Laptop.laptopCount += self.count
        
    @staticmethod
    def advantage():
        print("Laptops are portable")
        
    @classmethod
    def resetCount(cls):
        cls.laptopCount = 0


lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
# task 7

class Cat:
    Number_of_cats = 0
    
    def __init__(self, clr, act):
        self.clr = clr
        self.act = act
        Cat.Number_of_cats += 1
       
    @classmethod
    def no_parameter(cls):
        clr = "White"
        act = "sitting"
        return cls(clr, act)
    
    @classmethod
    def first_parameter(cls, clr):
        clr = clr
        act = "sitting"
        return cls(clr, act)
        
    @classmethod
    def second_parameter(cls, act):
        clr = "Grey"
        act = act
        return cls(clr, act)
    
    def changeColor(self, clr):
        self.clr = clr
    
    def printCat(self):
        print("{} cat is {}".format(self.clr, self.act))
    
    
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
# task 8

import math
class Cylinder:
    radius = 5
    height = 18
    
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        print("Default radius={} and height={}.".format(Cylinder.radius, Cylinder.height))
        Cylinder.radius = self.radius
        Cylinder.height = self.height
        print(f"Updated: radius={self.radius} and height={self.height}.")
        
    @classmethod
    def swap(cls, radius, height):
        swap = cls(height, radius)
        return swap
        
    @classmethod
    def changeFormat(cls, string):
        lst = string.split("-")
        objct = cls(float(lst[0]), float(lst[1]))
        return objct
    
    @staticmethod
    def area(radius, height):
        area = (2*math.pi*((radius)**2)) + (2*math.pi*radius*height)
        print("Area:", area)
    
    @staticmethod
    def volume(radius, height):
        volume = math.pi*(radius**2)*height
        print("Volume:", volume)



c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
# task 10

class SultansDine:
    total_branchs=0
    total_sell=0
    lst=[]
    index=0
    def __init__(self,location):
        self.location=location
        SultansDine.lst.append(self.location)
        self.sell=0
        SultansDine.total_branchs+=1
        
    def sellQuantity(self,sell):
        if sell < 10:
            self.sell = sell * 300
        elif sell < 20:
            self.sell = sell * 350
        else:
            self.sell = sell * 400
        SultansDine.total_sell+=self.sell
        SultansDine.lst.append(self.sell)
    
    def branchInformation(self):     
        print('Branch Name:',self.location)
        print('Branch Sell:',self.sell)
    
    @classmethod
    def details(cls):
        print('Total Number of branch(s):',cls.total_branchs)
        print('Total Sell:',cls.total_sell,'Taka')
        for i in range(SultansDine.total_branchs):
            
            print(f'Branch Name: {SultansDine.lst[SultansDine.index]}, Branch Sell: {SultansDine.lst[SultansDine.index+1]} taka')
            print(f"Branch consists of total sell's: {((SultansDine.lst[SultansDine.index+1]*100)/SultansDine.total_sell):.2f}%")
            SultansDine.index+=2
        SultansDine.index=0
            
#====================================================================
    
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
