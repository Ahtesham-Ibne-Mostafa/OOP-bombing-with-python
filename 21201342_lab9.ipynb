```python
#task1

class PlayerEarning:
    def __init__(self,a):
        self.name=a
    def calculateTotal(self,b,c=0):
        self.earn=b
        if c>30:
            self.bonus=int((5/100) * self.earn) + 10000
        elif c==0:
            self.bonus=0
        else:
            self.bonus=int((5/100) * self.earn) 
        self.total=self.earn+self.bonus
    def printDetails(self):
        print('Player Name:',self.name)
        print('Player Season Earning without bonus:',self.earn)
        print('Bonus:',self.bonus)
        print('Player Season Earning After Bonus:',self.total)

print("**********************")
player1 = PlayerEarning('Buffon')
player1.calculateTotal(250000)
player1.printDetails()

print("\n**********************")
player2 = PlayerEarning('Dybala')
player2.calculateTotal(250000, 31)
player2.printDetails()

print("\n**********************")
player3 = PlayerEarning('Cuadrado')
player3.calculateTotal(250000, 20)
player3.printDetails()
```

```python
# task 2

class myList:
    def __init__(self,*a):
        self.talika=list(a)
    def sum(self):
        self.total=sum(self.talika)
        print('Sum:',self.total)
    def merge(self,*b):
        self.talika+=list(b)
    def average(self):
        if self.talika==[]:
            print('Average:',0)
        else:
            print('Average:',self.total/len(self.talika))


l1 =  myList(2,3,4,5,6) #you might need a list inside your class to store the values
l1.sum()
l1.merge(4,5,9)
l1.sum()
l1.average()
print('----------------------------')
l2 =  myList()
l2.average()
l2.merge(1,2,4,8)
l2.sum()
```

```python
#task 3

class Bird:
    def __init__(self,a,b=False):
        self.name=a
        self.fy=b
        self.type='Flightless Birds'
    def fly(self):
        if self.fy==False:
            self.fly=str(self.name)+' can not fly'
        else:
            self.fly=str(self.name)+' can fly'
        print(self.fly)
    def setType(self,a):
        self.type=a
    def printDetail(self):
        print('Name:',self.name)
        print('Type:',self.type)
ostrich = Bird('Ostrich')
duck = Bird("Duck", True)
owl = Bird('Owl', True)
print('###########################')
ostrich.fly()
duck.fly()
owl.fly()
duck.setType('Water Birds')
owl.setType('Birds of Prey')
print('=========================')
ostrich.printDetail()
print('=========================')
duck.printDetail()
print('=========================')
owl.printDetail()
```

```python
# task 4

class Account:
    count=0
    def __init__(self,a,b,c,d):
        self.name=a
        self.age=b
        self.occ=c
        self.balance=d
        Account.count+=1
    def addMoney(self,a):
        self.balance+=a
    def withdrawMoney(self,a):
        if a <=self.balance:
            self.balance-=a
    def printDetails(self):
        print('Name:',self.name)
        print('Age:',self.age)
        print('Occupation:',self.occ)
        print('Total Amount:',self.balance)

print('No of account holders:', Account.count)
print("=========================")
p1 = Account("Abdul", 45, "Service Holder", 500000)
p1.addMoney(300000)
p1.printDetails()
print("=========================")
p2 = Account("Rahim", 55, "Businessman", 700000)
p2.withdrawMoney(700000)
p2.printDetails()
print("=========================")
p3 = Account("Ashraf", 62, "Govt. Officer", 200000)
p3.withdrawMoney(250000)
p3.printDetails()
print("=========================") 
print('No of account holders:', Account.count)
```

```python
# task 5

class Smartphone:
    def __init__(self,a=0):
        self.name=a
        self.subidha={}
    def addFeature(self,*a):
        if self.name!=0:
            if a[0] in self.subidha.keys():
                self.subidha[a[0]]+=[a[1]]
            elif a[0] not in self.subidha.keys() :
                self.subidha[a[0]]=[a[1]]
        else:
            print('Feature can not be added without phone name')
    def setName(self,a):
        self.name=a
    def printDetail(self):
        print('Phone Name:',self.name)

        for i in self.subidha.keys():
            print(i+': ',end='')
            x=''
            for j in self.subidha[i]:
                x+=j+', '
            print(x[0:-2])
    


s1 = Smartphone()
print("=================================")
s1.addFeature('Display', '6.1 inch')
print("=================================")
s1.setName('Samsung Note 20')
s1.addFeature('Display', '6.1 inch')
s1.printDetail()
print("=================================")
s2 = Smartphone('Iphone 12 Pro')
s2.addFeature('Display', '6.2 inch')
s2.addFeature('Ram', '6 GB')
print("=================================")
s2.printDetail()
s2.addFeature('Display', 'Amoled panel')
s2.addFeature('Ram', 'DDR5')
print("=================================")
s2.printDetail()
print("=================================")
```

```python
#task 8

class Processor:
    def __init__(self, model, thread, core):
        self.model = model
        self.core = core
        self.thread = thread
    def getInfo(self):
        return 'Model:'+self.model+ '\nCores:'+str(self.core)+ '\nThreads:'+ str(self.thread) 

class Intel(Processor):
    def __init__(self,a,b,c,d):
        Processor.__init__(self,a,c,b)
        self.price=d
    def getInfo(self):
        print (Processor.getInfo(self)+'\nPrice: '+str(self.price) )
class AMD(Processor):
    def __init__(self,a,b,c,d):
        Processor.__init__(self,a,c,b)
        self.price=d
    def getInfo(self):
        print (Processor.getInfo(self)+'\nPrice: '+str(self.price) )
      
p1 = Intel("Intel i5 10th Gen",6,12,17000)
p2 = AMD("Ryzen 5 3500X",6,6,13800)
p3 = AMD("Ryzen 5 3600",6,12,16900)
print('======================')
p1.getInfo()
print('======================')
p2.getInfo()
print('======================')
p3.getInfo()
```
