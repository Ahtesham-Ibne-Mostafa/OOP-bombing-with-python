```python
# task 1

class Calculator:
    def __init__(self,a,b):
        self.num1=a
        self.num2=b
        print('Let’s Calculate!')
    def add(self):
        self.result=self.num1+self.num2

    def subtract(self):
        self.result=self.num1-self.num2

    def multiply(self):
        self.result=self.num1*self.num2

    def divide(self):
        self.result=self.num1/self.num2

x=int(input('Enter the first value :'))
y=input('Enter the operator :')
z=int(input('Enter the second value :'))

calc=Calculator(x,z)
print('Value 1:',x)
print('Operator:',y)
print('Value 2:',z)

if y=='+':
    calc.add()
    print('Result :',calc.result)
elif y=='-':
    calc.subtract()
    print('Result :',calc.result)
elif y=='*':
    calc.multiply()
    print('Result :',calc.result)
elif y=='/':
    calc.divide()
    print('Result :',calc.result)
```

```python
# task 2

class Customer:
    def __init__(self,a):
        self.name=a
    def greet(self,b=None):
        if b==None:
            print('Hello!')
        else:
            print(f'Hello {self.name}!')
    def purchase(self,*c):
        print(f'{self.name}, you purchased {len(c)} item(s)')
        for i in c:
            print(i)

customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
# task 3

class Panda:
    def __init__(self,a,b,c):
        self.name=a
        self.gender=b
        self.age=c
    def sleep(self,y=None):
        if y==None:
            x=(f"{self.name}'s duration is unknown thus should have only bamboo leaves")
        elif 3<=y<=5:
            x=(f"{self.name} sleeps {y} hours daily and should have Mixed Veggies")
        elif 6<=y<=8:
            x=(f"{self.name} sleeps {y} hours daily and should have Eggplant & Tofu")
        elif 9<=y<=11:
            x=(f"{self.name} sleeps {y} hours daily and should have Broccoli Chicken")
        return x

panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
# task 4

class Cat:
  def __init__(self,color=None,state=None):
    if(color!=None and state!=None):
      self.color=color
      self.state=state
    elif(color==None and state!=None):
      self.state=state
    elif(color!=None and state==None):
      self.color=color
      self.state="sitting"
    else:
      self.color="White"
      self.state="sitting"
  def printCat(self):
    print(f"{self.color} cat is {self.state}")
  def changeColor(self,color):
    self.color=color
    print(f"{self.color} cat is {self.state}")
c1 = Cat()
c2 = Cat("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
```

```python
# task 5

class Student:
  def __init__(self,name=None):
    if(name!=None):
      self.name=name
    else:
      self.name="default student"
  def quizcalc(self,a,b=None,c=None):
    sum1=0
    if(b!=None and c!=None):
      sum1=sum1+a+b+c
    elif(b==None and c!=None):
      sum1=sum1+a+c
    elif(b!=None and c==None):
      sum1=sum1+a+b
    elif(b==None and c==None):
      sum1=sum1+a
    self.avg=sum1/3
  def printdetail(self):
    print("Hello",self.name)
    print("Your average quiz score is",self.avg)    
s1 = Student()
s1.quizcalc(10)
print('--------------------------------')
s1.printdetail()
s2 = Student('Harry')
s2.quizcalc(10,8)
print('--------------------------------')
s2.printdetail()
s3 = Student('Hermione')
s3.quizcalc(10,9,10)
```

```python
# task 9

class Patient:
  def __init__(self,name,age):
    self.name=name
    self.age=age
  def add_Symptom(self,a,b=None,c=None):
    if(b==None and c==None):
      self.a=a
      self.symptom=self.a
    elif(c==None):
      self.a=a
      self.b=b
      self.symptom=self.a+", "+self.b
    else:
      self.a=a
      self.b=b
      self.c=c
      self.symptom=self.a+", "+self.b+", "+self.c
  def printPatientDetail(self):
    print("Name: ",self.name)
    print("Age: ",self.age)
    print("Symptoms: ",self.symptom)
p1 = Patient("Thomas", 23)
p1.add_Symptom("Headache")
p2 = Patient("Carol", 20)
p2.add_Symptom("Vomiting", "Coughing")
p3 = Patient("Mike", 25)
p3.add_Symptom("Fever", "Headache", "Coughing")
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
print("=========================")
```

```python
# task 13

class Batsman :
    def __init__(self,b,c,a='New Batsman'):
        if a=='New Batsman':
            self.name=a
            self.run=b
            self.ball=c
        else:
            self.name=b
            self.run=c
            self.ball=a
    def setName(self,d):
        self.name=d
    def printCareerStatistics(self):
        print('Name:',self.name)
        print('Runs Scored:',self.run,',','Balls Faced:',self.ball)
    def battingStrikeRate(self):
        x=((self.run*100)/self.ball)
        return x

b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```
