```python
# task 6

z=input('Enter your new password :')
low, up, num, sign = 0, 0, 0, 0
error=''
for i in z:
    if i in "_$#@":
        sign+=1
    elif '0'<=i<='1':
        num+=1
    elif 'A'<= i <= 'Z':
        up+=1
    else:
        low+=1
if low!=0 and up!=0 and sign !=0 and num!=0 :
    error+='OK  '
else :
    if low==0:
        error+='Lowercase Missing, '
    if up==0:
        error+='Uppercase Missing, '
    if num==0:
        error+='Digit Missing, '
    if sign==0 :
        error+='Special Character Missing, '
print(error[0:-2])
```

Dictionary & Tuple

```python
# task 1
x=input('Enter the dictionary :').split(', ')
y=input('Enter the dictionary :').split(', ')
a={}
b={}
c=[]
for i in x:
    i=i.split(': ')
    a[i[0]]=int(i[1])
for i in y:
    i=i.split(': ')
    b[i[0]]=int(i[1])
for i in a.keys():
    if i in b.keys():
        b[i]=a[i]+b[i]
    else:
        b[i]=a[i]
for i in sorted(b):
    a[i]=b[i]
    if b[i] not in c:
        c.append(b[i])
print(a)
print('Values:',tuple(sorted(c)))
```

```python
# task 2

x={}
y=0
while y!=None:
    y=input('Enter the number :')
    if y== 'STOP':
        break
    elif y not in x.keys():
        x[y]=1
    else:
        x[y]=x[y]+1
for i in x.keys():
    print(i,'-',x[i],'times')
```

```python
# task 3

x=input('Enter the dictionary :').split(' , ')
a={}
for i in x:
    b=[]
    i=i.split(' : ')
    if i[1] not in a.keys():
        a[i[1]]=[i[0]]
    else:
        b.append(i[0])
        a[i[1]]=a[i[1]]+b
print(a)
```

```python
# task 4

x=input('Enter the string :')
y=input('Enter the string :')
z=1
p=len(y)
a={}
b={}
count=0
if len(x)!=len(y):
    print('Those strings are not anagrams.')
else:
    for i in x:
        a[i]=z
        z+=1
    for i in y:
        b[i]=p
        p-=1
    for i in a.keys():
        if i not in b.keys():
            print('Those strings are not anagrams.')
        elif a[i]==b[i]:
            count+=1
    if count==len(x) and count!=0:
        print('Those strings are anagrams.')
```
